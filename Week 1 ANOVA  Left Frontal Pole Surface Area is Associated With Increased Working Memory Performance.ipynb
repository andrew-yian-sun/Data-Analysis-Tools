# High Level Summary

[Zacharopoulos, Kingberg, and Kadosh, 2020](https://www.sciencedirect.com/science/article/pii/S0028393220301573?via%3Dihub) found that cortical surface area of the left frontal pole is negatively associated with working memory capacity, *specifically in the visuospatial modality*. This was in contrast with verbal working memory, which was not found to correlate with cortical surface area in this region. I wanted to see if how this association would hold up with another measure of working memory, the n-back task.

Therefore, I used data from the Human Connectome Project and ran an ANOVA to see if there were significant differences between n-back task accuracy across 4 levels (quartiles) of left frontal pole surface area. Interestingly, results showed that greater surface area was positively associated with accuracy on the n-back tasks.

## Analysis Plan and Hypothesis
I decided to perform an one-way between subjects ANOVA to examine the effect of brain structure on a working memory task. In particular, I wanted to see if and how the cortical surface area of the left frontal pole would be associated with mean differences in accuracy of the n-back task. 

Therefore, my $H_0$ would be that there are no differences in the nback accuracies for each surface area category, and my $H_1$ is that as surface area increases, accuracy for the n-back task will increase. 

If there are significant results, I will run a Tukey HSD test to examine which comparisons were significant.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import statsmodels.formula.api as smf # for the OLS module
import statsmodels.stats.multicomp as multi # for multiple comparisons test

# Mounting Google Drive onto Colab
import google.colab.drive as drive 
drive.mount('/content/gdrive')

# Reading the file
data = pd.read_csv('/content/gdrive/MyDrive/Data/HCP-behavioral.csv', low_memory=False)
```

## Converting from Continuous IV to Categorical IV
I first needed to turn the continous variable of surface area into a categorical variable, so I thought maybe separating surface area measurements into quartiles would work: https://towardsdatascience.com/what-are-quartiles-c3e117114cf1

```python
# Define quartiles
L_FP_SA = data[['FS_L_Frontalpole_Area', 'WM_Task_Acc']].dropna()
L_FP_SA = L_FP_SA['FS_L_Frontalpole_Area']
quartiles = np.quantile(L_FP_SA, .25), np.quantile(L_FP_SA, .50), np.quantile(L_FP_SA, .75)
# quartiles = np.quantile(L_FP_SA, .25, interpolation='midpoint'), np.quantile(L_FP_SA, .50, interpolation='midpoint'), np.quantile(L_FP_SA, .75, interpolation='midpoint')

# look at the distribution
plt.style.use('seaborn-talk')
plt.hist(L_FP_SA)
for i in quartiles:
  plt.axvline(i, color='r')
plt.xlabel('Surface area ($mm^2$)')
plt.title('Histogram of Left Frontal Pole Surface Area')
plt.show()
# print(f'Quartiles are: \n25%tile: {quartiles[0]}mm, 50%tile: {quartiles[1]}mm, 75%tile: {quartiles[2]}mm'.format(quartiles))
L_FP_SA.describe()
```

Conditional column creation: https://www.delftstack.com/howto/python-pandas/how-to-create-dataframe-column-based-on-given-condition-in-pandas/

```python
# np.select requires a list for your selection conditions and a list for your choice conditions
# the indices should be matched for both the selection and choice conditions
conditionlist = [
                 (data['FS_L_Frontalpole_Area'] < quartiles[0]),
                 (data['FS_L_Frontalpole_Area'] >= quartiles[0]) & (data['FS_L_Frontalpole_Area'] < quartiles[1]),
                 (data['FS_L_Frontalpole_Area'] >= quartiles[1]) & (data['FS_L_Frontalpole_Area'] < quartiles[2]),
                 (data['FS_L_Frontalpole_Area'] >= quartiles[2]),
]
choicelist = ['1st', '2nd', '3rd', '4th']

data['FS_L_Frontalpole_Area_Quartile'] = np.select(conditionlist, choicelist, default=None)
```

## Subsetting the Data

```python
# I've included the continuous surface area variable as well to double check that the quartile conversion was correct
leftFP_WM_df = data[['FS_L_Frontalpole_Area', 'FS_L_Frontalpole_Area_Quartile', 'WM_Task_Acc']]
leftFP_WM_df_ = leftFP_WM_df.dropna()
leftFP_WM_df_
```

# Analysis

## ANOVA Table

```python
model1= smf.ols(formula='WM_Task_Acc ~ C(FS_L_Frontalpole_Area_Quartile)', data=leftFP_WM_df_)
print(model1.fit().summary())
```

## Group Means

```python
leftFP_WM_df_counts = leftFP_WM_df_.groupby('FS_L_Frontalpole_Area_Quartile').count()
print('\nGroup Counts', leftFP_WM_df_counts, sep='\n')
leftFP_WM_df_means = leftFP_WM_df_.groupby('FS_L_Frontalpole_Area_Quartile').mean()
print('\nGroup Means', leftFP_WM_df_means, sep='\n')
leftFP_WM_df_stds = leftFP_WM_df_.groupby('FS_L_Frontalpole_Area_Quartile').std()
print('\nGroup Stds', leftFP_WM_df_stds, sep='\n')
```

Bar plotting: https://benalexkeen.com/bar-charts-in-matplotlib/

```python
# defining variables for bar plot
x = list(leftFP_WM_df_means.index)
x_pos = [i for i, _ in enumerate(x)]
y = leftFP_WM_df_means.values[:, 1]
yerr = leftFP_WM_df_stds.values[:, 1] # error of y
print(x_pos, x, y, yerr, sep='\n')
```

```python
plt.style.use('seaborn-talk')
plt.bar(x=x_pos, height=y, yerr=yerr, color=[(0.2, 0.2, 0.8, 0.25), (0.2, 0.2, 0.8, 0.50), (0.2, 0.2, 0.8, 0.75), (0.2, 0.2, 0.8, 0.75)])
plt.xticks(x_pos, x)
plt.ylim(60, 100)
plt.xlabel('Quartiles')
plt.ylabel('Mean Surface Area ($mm^2$)')
plt.show()
```

## Multiple Comparisons

```python
mc1 = multi.MultiComparison(leftFP_WM_df_['WM_Task_Acc'], leftFP_WM_df_['FS_L_Frontalpole_Area_Quartile'])
print(mc1.tukeyhsd().summary())
```

# Write-Up

## APA Style
A one-way between subjects ANOVA was conducted to compare the effect of left frontal pole surface area (mm) on mean accuracy (% correct) of a working memory n-back task. Left frontal pole surface area was separated into low, medium-low, medium-high, and high conditions (quartiles). 

Results showed a significant effect of left frontal pole surface area on mean accuracy of a working memory n-back task at the p<.05 level for the four conditions [F(3, 1087) = 5.055, p = 0.002]. 

Post hoc comparisons using the Tukey HSD test indicated that the mean accuracy for the low condition (M = 85.12, SD = 9.46) was significantly different than the medium-high (M = 87.47, SD = 8.82) and high conditions (M = 87.92, SD = 9.59). However, the low-medium condition (M = 86.15, SD = 9.34) was not significantly different from any other condition, and neither was the comparison between the high-medium and high conditions.

## Plain English
I saw a relatively recent neuroscience paper suggesting that a frontal region of the brain was negatively associated with visualspatial working memory performance, which is important for many tasks. For example, measures of working memory capacity are predictive of things like IQ. 

I wanted to see if those recent results would be replicated with another working memory task, so I downloaded data from the Human Connectome Project (which is separate from the recent paper) and ran a statistical test, the results of which suggested that in contrast to the findings of the paper, this particular frontal region is positively associated with this measure of working memory.

## Discussion
Unfortunately, because the task used in the Zacharopoulos study was not the same as the one used in the HCP dataset, this wasn't a direct replication. However, if the conclusions of both these analyses are accurate, there's an interesting question as to why the relationship between left frontal pole surface area flips with these 2 different modalities. Zacharopoulos et al make it clear that the association they find is specific only to visuospatial working memory, and not another measure like verbal working memory (digit span), which did not correlate with left frontal pole surface area in their study.

Further questions:
1. **Do other regions have similar associations?:** The HCP dataset includes 34 areas on each hemisphere of the brain that have surface area measurements, as well as additional measurements for cortical thickness and volume, so further analysis with this method could easily be scripted.
2. **Is this region specific for working memory?:** Because working memory is a higher-order function, it would be unsurprising to see this region associated with other cognitive functions.
3. **Is this a causal relationship?:** Maybe a TMS study could help determine whether this region is necessary for working memory.
